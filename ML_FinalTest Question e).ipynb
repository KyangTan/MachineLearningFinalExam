```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset
dataset = pd.read_csv("C:\\Users\\kwany\\Desktop\\ML Final Test\\question1edata.csv")
print("Original Feature Vector")
print(dataset.to_numpy())
```

```python
# Features standardization
sc = StandardScaler()
std_dataset = sc.fit_transform(dataset)
std_dataset
```

```python
# Construct the covariance matrix
covariance_matrix = np.cov(std_dataset.T)
print("Covariance Matrix")
print(covariance_matrix)
```

```python
# Get the eigenvectors and eigenvalues by using the covariance matrix
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
print("Eigenvalues")
print(eigen_values)
print("")
print("Eigenvector")
print(eigen_vectors)
```

```python
# Obtain the feature vector using eigenvectors
feature_vec = np.dot(std_dataset, eigen_vectors)
print("Feature Vector")
print(feature_vec)
```

```python
# Calculate cumulative sum of explained variances
tot = sum(eigen_values)
var_exp = [(i / tot) for i in sorted(eigen_values, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
```
